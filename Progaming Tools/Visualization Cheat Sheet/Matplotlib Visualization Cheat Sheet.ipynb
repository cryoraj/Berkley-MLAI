# Matplotlib Visualization Cheat Sheet

ðŸ”¹ Distribution Plots
- hist: Frequency distribution â†’ plt.hist()
- box: Summary statistics â†’ plt.boxplot()

ðŸ”¹ Relational Plots
- scatter: Relationship between variables â†’ plt.scatter()
- line: Trends over time â†’ plt.plot()

ðŸ”¹ Categorical Plots
- bar: Category totals or means â†’ plt.bar()
- barh: Horizontal bar chart â†’ plt.barh()
- pie: Proportional breakdown â†’ plt.pie()

ðŸ”¹ Matrix Plots
- imshow: Display 2D arrays â†’ plt.imshow()
- matshow: Matrix visualization â†’ plt.matshow()

ðŸ”¹ Regression Plots
- scatter + line: Manual regression overlay â†’ plt.scatter() + plt.plot()

ðŸ”¹ Styling & Annotation
- title: Add plot title â†’ plt.title()
- xlabel / ylabel: Axis labels â†’ plt.xlabel(), plt.ylabel()
- legend: Add legend â†’ plt.legend()
- grid: Add gridlines â†’ plt.grid(True)
- savefig: Export figure â†’ plt.savefig('filename.png', dpi=300)

ðŸ”¹ Subplots
- subplot: Single row/column layout â†’ plt.subplot()
- subplots: Grid layout â†’ fig, ax = plt.subplots()
- sharex/sharey: Shared axes â†’ plt.subplots(sharex=True)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# -------------------------------
# ðŸ”§ Dummy Data Generator
# -------------------------------
def generate_dummy_data(n=500, seed=42):
    np.random.seed(seed)
    df = pd.DataFrame({
        'AcceptedCoupon': np.random.binomial(1, 0.4, n),
        'Age': np.random.normal(40, 10, n).astype(int),
        'Income': np.random.normal(60000, 15000, n).astype(int),
        'MaritalStatus': np.random.choice(['Single', 'Married', 'Divorced'], n),
        'PassengerType': np.random.choice(['Business', 'Personal', 'Family'], n),
        'TripDuration': np.random.exponential(scale=3, size=n),
        'CouponType': np.random.choice(['Restaurant', 'Grocery', 'Retail'], n)
    })
    return df

# -------------------------------
# ðŸ“Š Visualization Functions
# -------------------------------
def plot_histogram(df, column):
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_box(df, column):
    plt.figure(figsize=(8, 5))
    plt.boxplot(df[column], vert=False)
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_scatter(df, x, y):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x], df[y], alpha=0.6, c='teal')
    plt.title(f'{y} vs {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_line(df, x, y):
    plt.figure(figsize=(8, 5))
    sorted_df = df.sort_values(by=x)
    plt.plot(sorted_df[x], sorted_df[y], marker='o', linestyle='-', color='purple')
    plt.title(f'{y} over {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_bar(df, category, value):
    agg = df.groupby(category)[value].mean()
    plt.figure(figsize=(8, 5))
    plt.bar(agg.index, agg.values, color='coral')
    plt.title(f'Mean {value} by {category}')
    plt.xlabel(category)
    plt.ylabel(f'Mean {value}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_pie(df, category):
    counts = df[category].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Distribution of {category}')
    plt.tight_layout()
    plt.show()

def plot_heatmap(df):
    corr = df.corr(numeric_only=True)
    plt.figure(figsize=(10, 8))
    plt.imshow(corr, cmap='coolwarm', interpolation='none')
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()

def plot_regression(df, x, y):
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x], df[y], alpha=0.5, label='Data')
    m, b = np.polyfit(df[x], df[y], 1)
    plt.plot(df[x], m * df[x] + b, color='red', label='Regression Line')
    plt.title(f'Regression: {y} vs {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

```

```python
# -------------------------------
# ðŸš€ Sample Execution
# -------------------------------
if __name__ == "__main__":
    df = generate_dummy_data()

    print("ðŸ”¹ Histogram: Age Distribution")
    plot_histogram(df, 'Age')

    print("ðŸ”¹ Boxplot: Income")
    plot_box(df, 'Income')

    print("ðŸ”¹ Scatter Plot: Income vs Age")
    plot_scatter(df, 'Age', 'Income')

    print("ðŸ”¹ Line Plot: Income over Age")
    plot_line(df, 'Age', 'Income')

    print("ðŸ”¹ Bar Plot: Mean Coupon Acceptance by Type")
    plot_bar(df, 'CouponType', 'AcceptedCoupon')

    print("ðŸ”¹ Pie Chart: Passenger Type Distribution")
    plot_pie(df, 'PassengerType')

    print("ðŸ”¹ Heatmap: Correlation Matrix")
    plot_heatmap(df)

    print("ðŸ”¹ Regression Plot: Income vs Age")
    plot_regression(df, 'Age', 'Income')
```
