```python
"""
Seaborn Visualization Cheat Sheet

ðŸ”¹ Distribution Plots
- histplot: Frequency distribution â†’ sns.histplot()
- kdeplot: Density estimation â†’ sns.kdeplot()
- ecdfplot: Cumulative distribution â†’ sns.ecdfplot()
- rugplot: Individual observations â†’ sns.rugplot()

ðŸ”¹ Relational Plots
- scatterplot: Relationship between variables â†’ sns.scatterplot()
- lineplot: Trends over time â†’ sns.lineplot()

ðŸ”¹ Categorical Plots
- boxplot: Distribution + outliers â†’ sns.boxplot()
- violinplot: Distribution shape + density â†’ sns.violinplot()
- stripplot: Jittered data points â†’ sns.stripplot()
- swarmplot: Non-overlapping points â†’ sns.swarmplot()
- barplot: Mean comparisons â†’ sns.barplot()
- countplot: Category frequency â†’ sns.countplot()

ðŸ”¹ Matrix Plots
- heatmap: Correlation/tabular data â†’ sns.heatmap()
- clustermap: Hierarchical clustering â†’ sns.clustermap()

ðŸ”¹ Multi-Plot Grids
- FacetGrid: Subplots by category â†’ sns.FacetGrid()
- pairplot: Pairwise relationships â†’ sns.pairplot()
- jointplot: Scatter + marginal distributions â†’ sns.jointplot()

ðŸ”¹ Regression Plots
- regplot: Linear regression â†’ sns.regplot()
- lmplot: Regression + faceting â†’ sns.lmplot()
"""

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Global styling
sns.set_theme(style='whitegrid')

# -------------------------------
# ðŸ“Š Plotting Functions
# -------------------------------

def plot_histogram(df, column, bins=30, kde=True, title=None):
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=column, bins=bins, kde=kde)
    plt.title(title or f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

def plot_kde(df, column, title=None):
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data=df, x=column, fill=True)
    plt.title(title or f'Density of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.tight_layout()
    plt.show()

def plot_scatter(df, x, y, hue=None, title=None):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x=x, y=y, hue=hue)
    plt.title(title or f'{y} vs {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()

def plot_line(df, x, y, title=None):
    plt.figure(figsize=(8, 5))
    sns.lineplot(data=df, x=x, y=y, marker='o')
    plt.title(title or f'{y} over {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()

def plot_box(df, x, y, title=None):
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x=x, y=y)
    plt.title(title or f'{y} Distribution by {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()

def plot_violin(df, x, y, title=None):
    plt.figure(figsize=(8, 5))
    sns.violinplot(data=df, x=x, y=y, inner='quartile')
    plt.title(title or f'{y} Density by {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()

def plot_bar(df, x, y, title=None):
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df, x=x, y=y, ci='sd')
    plt.title(title or f'Mean {y} by {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()

def plot_count(df, x, title=None):
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=x, order=df[x].value_counts().index)
    plt.title(title or f'Count of {x}')
    plt.xlabel(x)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

def plot_heatmap(corr_matrix, title='Correlation Heatmap'):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title(title)
    plt.tight_layout()
    plt.show()

def plot_pairplot(df, hue=None, diag_kind='kde'):
    sns.pairplot(df, hue=hue, diag_kind=diag_kind)
    plt.tight_layout()
    plt.show()

def plot_jointplot(df, x, y, kind='reg'):
    sns.jointplot(data=df, x=x, y=y, kind=kind)
    plt.tight_layout()
    plt.show()

def plot_regression(df, x, y, title=None):
    plt.figure(figsize=(8, 5))
    sns.regplot(data=df, x=x, y=y, line_kws={'color': 'red'})
    plt.title(title or f'Regression: {y} vs {x}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()

def plot_lm(df, x, y, hue=None, col=None, title=None):
    sns.lmplot(data=df, x=x, y=y, hue=hue, col=col)
    plt.tight_layout()
    plt.show()



"""
Seaborn Visualization Cheat Sheet + Reusable Plotting Functions

Includes:
- Distribution, relational, categorical, matrix, regression plots
- Best practices for reproducibility
- Dummy dataset for testing
"""

# -------------------------------
# ðŸ”§ Dummy Data Generator
# -------------------------------
def generate_dummy_data(n=500, seed=42):
    np.random.seed(seed)
    df = pd.DataFrame({
        'AcceptedCoupon': np.random.binomial(1, 0.4, n),
        'Age': np.random.normal(40, 10, n).astype(int),
        'Income': np.random.normal(60000, 15000, n).astype(int),
        'MaritalStatus': np.random.choice(['Single', 'Married', 'Divorced'], n),
        'PassengerType': np.random.choice(['Business', 'Personal', 'Family'], n),
        'TripDuration': np.random.exponential(scale=3, size=n),
        'CouponType': np.random.choice(['Restaurant', 'Grocery', 'Retail'], n)
    })
    return df




```

```python
if __name__ == "__main__":
    df = generate_dummy_data()

    print("ðŸ”¹ Histogram: Age Distribution")
    plot_histogram(df, column='Age', title='Age Distribution')

    print("ðŸ”¹ KDE Plot: Income Density")
    plot_kde(df, column='Income', title='Income Density')

    print("ðŸ”¹ Scatter Plot: Income vs Age by Marital Status")
    plot_scatter(df, x='Age', y='Income', hue='MaritalStatus', title='Income vs Age by Marital Status')

    print("ðŸ”¹ Boxplot: Income by Marital Status")
    plot_box(df, x='MaritalStatus', y='Income', title='Income by Marital Status')

    print("ðŸ”¹ Violin Plot: Trip Duration by Passenger Type")
    plot_violin(df, x='PassengerType', y='TripDuration', title='Trip Duration by Passenger Type')

    print("ðŸ”¹ Bar Plot: Coupon Acceptance by Type")
    plot_bar(df, x='CouponType', y='AcceptedCoupon', title='Coupon Acceptance by Type')

    print("ðŸ”¹ Count Plot: Passenger Type Frequency")
    plot_count(df, x='PassengerType', title='Passenger Type Frequency')

    print("ðŸ”¹ Heatmap: Correlation Matrix")
    plot_heatmap(df.corr(numeric_only=True))

    print("ðŸ”¹ Pairplot: Age, Income, TripDuration, AcceptedCoupon")
    plot_pairplot(df[['Age', 'Income', 'TripDuration', 'AcceptedCoupon']], hue='AcceptedCoupon')

    print("ðŸ”¹ Jointplot: Income vs Age")
    plot_jointplot(df, x='Age', y='Income')

    print("ðŸ”¹ Regression Plot: Income vs Age")
    plot_regression(df, x='Age', y='Income')

    print("ðŸ”¹ LM Plot: Income vs Age by Marital Status and Passenger Type")
    plot_lm(df, x='Age', y='Income', hue='MaritalStatus', col='PassengerType')
```
